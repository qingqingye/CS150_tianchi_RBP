````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import Imputer
%matplotlib inline
````

````python
TRAIN_CSV = 'train.csv'
TEST_CSV = 'test.csv'
````

````python
train_data = pd.read_csv(TRAIN_CSV)
test_data = pd.read_csv(TEST_CSV)
````

# Basic info

````python
train_data.info()
````

````python
train_data.head()
````

````python
train_data = train_data.set_index(['Id'])
````

````python
train_data.head()
````

````python
train_data.isnull().sum()
````

````python
train_data.LotFrontage.head() #continuous
````

````python
train_data.FireplaceQu.head()
````

````python
imputer_mean = Imputer(strategy='mean')
train_data.LotFrontage = imputer_mean.fit_transform(train_data.LotFrontage.values[:,np.newaxis])
````

````python
train_data.LotFrontage.isnull().sum()
````

````python
train_data.FireplaceQu.loc[train_data.FireplaceQu.isnull()] = 'Nan'
````

````python
train_data.FireplaceQu.unique()
````

````python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(train_data.FireplaceQu.unique())
````

````python
le.classes_
````

````python
train_data.FireplaceQu = le.transform(train_data.FireplaceQu)
````

````python
train_data.FireplaceQu[train_data.FireplaceQu==3] = np.NAN
````

````python
train_data.FireplaceQu.isnull().sum()
````

````python
imputer_most_frequent = Imputer(strategy='most_frequent')
train_data.FireplaceQu = imputer_most_frequent.fit_transform(train_data.FireplaceQu[:,np.newaxis])
````

# Observing distribution

````python
import seaborn as sns
# y = train['SalePrice']
import scipy.stats as st
y = train_data['SalePrice']
plt.figure(1); plt.title('Log Normal')
sns.distplot(y, kde=False, fit=st.lognorm)
````

````python
plt.figure()
sns.distplot(np.log1p(y), kde=False)
````

# Relation between different variable

````python
#scatter plot grlivarea/saleprice
var = 'GrLivArea'
data = pd.concat([train_data['SalePrice'], train_data[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
````

````python
#box plot overallqual/saleprice
var = 'OverallQual'
data = pd.concat([train_data['SalePrice'], train_data[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
````

````python
#correlation matrix
corrmat = train_data.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
````

````python
#saleprice correlation matrix
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(train_data[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
````
